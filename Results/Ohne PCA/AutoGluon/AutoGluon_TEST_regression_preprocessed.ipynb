```python
import autogluon
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor 
```

```python
train_data = TabularDataset('train_reg_df.csv')
train_data.head()
```

```python
label = 'Price'
train_data[label].describe()
```

```python
predictor_price = TabularPredictor(label=label, path="agModels_perdict_price_normal_TEST").fit(
    train_data, 
    time_limit=14400, 
    presets='best_quality',
    ag_args_fit={'num_cpus': 16}
)
```

```python
test_data = TabularDataset('test_reg_df.csv')
```

```python
predictor_price.evaluate(test_data)
```

```python
predictor_price.leaderboard(test_data)
```

```python
import pandas as pd

leaderboard_df_reg_normal = predictor_price.leaderboard(test_data, extra_info=True)

print(leaderboard_df_reg_normal)
```

```python
from IPython.display import FileLink
leaderboard_df_reg_normal.to_csv('autogluon_TEST_leaderboard_regression_normal.csv', index=False)
FileLink('autogluon_TEST_leaderboard_regression_normal.csv')
```
