```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV files from training
leaderboard_normal = pd.read_csv('autogluon_leaderboard_regression_normal.csv')
leaderboard_preprocessed = pd.read_csv('autogluon_leaderboard_regression_preprocessed.csv')

#top 5 models from the original dataset
top_models_normal = leaderboard_normal.head(5)
top_models_preprocessed = leaderboard_preprocessed.head(5)

# Set up the plotting
plt.figure(figsize=(10, 6))

#test scores for both training results side by side for each model (top 5)
bar_width = 0.35
index = top_models_normal.index

plt.bar(index, top_models_normal['score_test'], bar_width, label='Original Data', color='blue')
plt.bar(index + bar_width, top_models_preprocessed['score_test'], bar_width, label='Preprocessed Data', color='green')

# Adding labels and title
plt.xlabel('Model')
plt.ylabel('Test Score')
plt.title('Test Score Comparison: Original vs Preprocessed Data')
plt.xticks(index + bar_width / 2, top_models_normal['model'], rotation=45, ha="right")
plt.legend()

plt.tight_layout()
plt.show()
```
