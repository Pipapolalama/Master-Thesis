```python
import pandas as pd
import matplotlib.pyplot as plt

#CSV files from training
metrics_normal = pd.read_csv('autogluon_regression_predictorScore_normal.csv')
metrics_preprocessed = pd.read_csv('autogluon_regression_predictorScore_preprocessed.csv')

#RSME value as metric of interest
rmse_normal = metrics_normal['root_mean_squared_error'].iloc[0]
rmse_preprocessed = metrics_preprocessed['root_mean_squared_error'].iloc[0]

# Data to plot
variants = ['Original Data', 'Preprocessed Data']
rmse_values = [rmse_normal, rmse_preprocessed]

#bar chart for comparising
plt.figure(figsize=(8, 6))
plt.bar(variants, rmse_values, color=['blue', 'green'])

plt.xlabel('Data Variant')
plt.ylabel('Root Mean Squared Error (RMSE)')
plt.title('RMSE Comparison: Original vs Preprocessed Data')


plt.tight_layout()
plt.savefig('AutoGluon_comparison_regression_rmse.png', dpi=300)
plt.show()
```
