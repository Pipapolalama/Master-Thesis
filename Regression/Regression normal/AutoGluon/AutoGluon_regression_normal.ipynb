```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
from sklearn.model_selection import train_test_split
dataset = TabularDataset('combined.csv')

dataset = dataset.dropna(subset=['Price'])  # Remove rows with missing prices

df_train, df_test = train_test_split(dataset, test_size=0.25)
```

```python
train_data = df_train
train_data.head()
```

```python
test_data = df_test
test_data.head()
```

```python
price_column = 'Price'
train_data[price_column].head()
```

```python
#4h train-time
predictor_price = TabularPredictor(label=price_column, path="agModels_perdict_price").fit(train_data, time_limit=14400, presets='best_quality')
```

```python
predictor_price.evaluate(test_data)
```

```python
predictor_price.leaderboard(test_data)
```

```python
#leaderbaord download
import pandas as pd

leaderboard_df = predictor_price.leaderboard(test_data, extra_info=True)

print(leaderboard_df)
```

```python
from google.colab import files

leaderboard_df.to_csv('autogluon_leaderboard_regression_normal.csv', index=False)

files.download('autogluon_leaderboard_regression_normal.csv')
```

```python
#predictor scores download
import pandas as pd
from google.colab import files

#predict score
evaluation_results = predictor_price.evaluate(test_data)

#to df
evaluation_df = pd.DataFrame([evaluation_results])

print(evaluation_df)

#to CSV
evaluation_df.to_csv('autogluon_regression_predictorScore_normal.csv', index=False)


files.download('autogluon_regression_predictorScore_normal.csv')
```
